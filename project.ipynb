```python
Total_proste=int(Rel_b['Pentecostal Protestant Christian']*Total_population)
Total_non_proste=int(Rel_b['Non Pentecostal Protestant Christian']*Total_population)
Total_Islam=int(Rel_b['Islam']*Total_population)
Total_Roman=int(Rel_b['Roman Catholic Christian']*Total_population)
Total_Other_christian=int(Rel_b['Other Christians']*Total_population)
Total_trad=int(Rel_b['Traditional African Beliefs']*Total_population)
Total_atheism=int(Rel_b['Atheism or Agnosticism']*Total_population)
Total_other_religion=int(Rel_b['Rastafari, Buddhism, Hinduism, and other Beliefs']*Total_population)
```

```python
bar_plot_data=[Total_proste,Total_non_proste,Total_Islam,Total_Roman,Total_Other_christian,Total_trad,Total_atheism,Total_other_religion]
#bar_plot_data=np.array(bar_plot_data)

bar_plot_xlabels=['Protestant','Non Protestant','Islam','Catholic','Other Christians','Traditionalist','Atheism','Other Religion']
bar_plot_data=pd.DataFrame(bar_plot_data)
bar_plot_data
```

```python
import matplotlib.pyplot as plt
import math
from scipy import stats
%matplotlib inline


col=('Protestant','Non Protestant','Islam','Catholic','Other Christians','Traditionalist','Atheism','Other Religion')
x_pos=np.arange(len(col))
maxi=[Total_proste,Total_non_proste,Total_Islam,Total_Roman,Total_Other_christian,Total_trad,Total_atheism,Total_other_religion]
maxi
```

```python
plt.figure()
plt.bar(x_pos,maxi,color=['blue','red','green','black','yellow','violet','pink','red'],
ec=['black','black','black','green','black','black','black','black'])
plt.title('A Graph of Religious people in Ghana ')
plt.xticks(x_pos,col)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Religion people in Ghana')
plt.ylabel('Number of religious people in millions')
plt.legend(['Protestant','Non_Protestant','Islam','Catholic','Other Christians','Traditionalist','Atheism','Other Religion'])
for spine in plt.gca().spines.values():
    spine.set_visible(False)
```

```python
labels=['Protestant','Non_Protestant','Islam','Catholic','Other Christians','Traditionalist','Atheism','Other Religion']
plt.pie(maxi,labels=labels)
plt.title('A Piechart of religious people in Ghana')
plt.legend(labels)
```

```python
celebrate_christmas=Total_proste+Total_non_proste+Total_Roman+Total_Other_christian
christ_mas=[celebrate_christmas,Total_Islam,Total_other_religion,Total_atheism,Total_trad]
plt.pie(christ_mas,labels=['Christmas','Muslims','Other Religion','Atheist','Tradionalist'])
plt.title('Piechart of Christmas Celebrants in Ghana')
plt.legend(['Christians','Muslims','Other Religion','Atheist','Traditionalist'])
```

```python
print('There are {} christmas celebrants in Ghana'.format(celebrate_christmas))
```
